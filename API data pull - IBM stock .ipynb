```python
def api_runner():
    global df 
    
    url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=60min&apikey=JYKKBEC9UDEDM95G'
    r = requests.get(url)
    data = r.json()
    
    df2 = pd.json_normalize(data)
    df2['Timestamp'] = pd.to_datetime('now')
    df = df.append(df2)
```
